# 1. Definiciones

```

def parse_complex(s):
    """Parse a string like '0.5+0.3j' into a complex number."""
    return complex(s.replace('i', 'j'))

def make_unitary(entries):
    """Build a 2x2 matrix from a list of 4 complex entries (row-major)."""
    return np.array(entries, dtype=complex).reshape(2, 2)

def kronecker(U1, U2):
    """Compute the tensor product U1 ⊗ U2."""
    return np.kron(U1, U2)

def apply_to_rho(U, rho):
    """Apply U to rho: rho' = U rho U†"""
    return U @ rho @ U.conj().T

def check_properties(rho, label="ρ'"):
    print(f"\n--- Properties of {label} ---")
    print(f"  Trace:       {np.trace(rho).real:.6f}")
    print(f"  Hermitian:   {np.allclose(rho, rho.conj().T)}")
    print(f"  Purity Tr(ρ²): {np.trace(rho @ rho).real:.6f}")
    eigenvalues = np.linalg.eigvalsh(rho)
    print(f"  Eigenvalues: {np.round(eigenvalues, 6)}")
    print(f"  PSD (all eigenvalues ≥ 0): {np.all(eigenvalues >= -1e-9)}")
```

# 2.1.3 Code to obtain different Rotations to every matrix (and check their negativites)

```
#Transformar matrices y numero de matrices negativas
import numpy as np

def random_SU2():
    t1, t2, t3, t4 = np.random.uniform(0, 2*np.pi, size=4)
    U = np.array([
        [np.exp(1j*(t1 - t2/2 - t3/3)) * np.cos(t4/2),
         -np.exp(1j*(t1 - t2/2 + t3/3)) * np.sin(t4/2)],
        [np.exp(1j*(t1 + t2/2 - t3/3)) * np.sin(t4/2),
         np.exp(1j*(t1 + t2/2 + t3/3)) * np.cos(t4/2)]
    ], dtype=complex)
    return U

def partial_transpose(rho, dim=2):
    d = dim
    rho_pt = np.zeros_like(rho)
    for i in range(d):
        for j in range(d):
            for k in range(d):
                for l in range(d):
                    rho_pt[i*d+k, j*d+l] = rho[i*d+l, j*d+k]
    return rho_pt

neg_limit = 2e-9

def negativity(rho):
    rho_pt    = partial_transpose(rho)
    eigenvals = np.linalg.eigvalsh(rho_pt)
    return np.sum(np.abs(eigenvals[eigenvals < -neg_limit]))

def parse_line_to_rho(line):
    nums = list(map(float, line.strip().split()))
    if len(nums) != 32:
        raise ValueError(f"Expected 32 numbers per line, got {len(nums)}")
    complexes = [complex(nums[2*k], nums[2*k+1]) for k in range(16)]
    return np.array(complexes, dtype=complex).reshape(4, 4)

def apply_to_rho(U, rho):
    return U @ rho @ U.conj().T

def check_properties(rho, rho_out, index):
    tr      = np.trace(rho_out).real
    purity  = np.trace(rho_out @ rho_out).real
    herm    = np.allclose(rho_out, rho_out.conj().T)
    eigvals = np.linalg.eigvalsh(rho_out)
    psd     = np.all(eigvals >= -1e-9)
    neg_in  = negativity(rho)
    neg_out = negativity(rho_out)
    print(f"  ρ'[{index}] — Tr={tr:.4f}  Purity={purity:.4f}  "
          f"Neg(ρ)={neg_in:.4f}  Neg(ρ')={neg_out:.4f}  "
          f"Hermitian={herm}  PSD={psd}")
    return neg_in, neg_out   # return both so the loop can count them

#input_file =  "input_data/T8/T8_(0.0, 0.1).txt"
input_file =  "input_data/T4/mixed_states/negativity_(0.0, 0.1).txt"
#input_file =  "input_data/T3/pure_states/negativity_(0.0, 0.1).txt"
output_file = "input_data/trash.txt"

with open(input_file, "r") as f:
    lines = [l for l in f.readlines() if l.strip() and not l.strip().startswith("#")]

print(f"Found {len(lines)} matrices.\n")

count_zero_before = 0
count_zero_after  = 0

with open(output_file, "w") as fout:
    for idx, line in enumerate(lines):
        U1 = random_SU2()
        U2 = random_SU2()
        U  = np.kron(U1, U2)
        rho     = parse_line_to_rho(line)
        rho_out = apply_to_rho(U, rho)
        neg_in, neg_out = check_properties(rho, rho_out, idx)
        # --- Increment counters if negativity is zero ---
        if neg_in  < neg_limit:
            count_zero_before += 1
        if neg_out < neg_limit:
            count_zero_after  += 1

        flat = rho_out.flatten()
        nums = []
        for c in flat:
            nums.append(f"{c.real:.10f}")
            nums.append(f"{c.imag:.10f}")
        fout.write("\t".join(nums) + "\n")
# --- Summary ---
print(f"\n--- Summary ---")
print(f"Total matrices:               {len(lines)}")
print(f"Zero negativity before U:     {count_zero_before}")
print(f"Zero negativity after  U:     {count_zero_after}")
print(f"\nFile origin is {input_file}")
print(f"\nResults saved to {output_file}")
```

# 3 Code to transform matrices to features parameters

```
import numpy as np

def partial_trace_A(rho):
    """Trace out qubit A (first qubit), returns 2x2 reduced state on B"""
    return np.array([[rho[0,0]+rho[2,2], rho[0,1]+rho[2,3]],
                     [rho[1,0]+rho[3,2], rho[1,1]+rho[3,3]]])

def partial_trace_B(rho):
    """Trace out qubit B (second qubit), returns 2x2 reduced state on A"""
    return np.array([[rho[0,0]+rho[1,1], rho[0,2]+rho[1,3]],
                     [rho[2,0]+rho[3,1], rho[2,2]+rho[3,3]]])

def partial_transpose_B(rho):
    """Partial transpose with respect to qubit B"""
    rho_pt = np.zeros_like(rho)
    for i in range(2):
        for j in range(2):
            for k in range(2):
                for l in range(2):
                    rho_pt[i*2+l, k*2+j] = rho[i*2+j, k*2+l]
    return rho_pt

def lu_invariants(rho):
    """
    Compute 9 LU-invariant features from a 4x4 density matrix:
      [t2, t3, t4, a2, b2, λ1, λ2, λ3, λ4]
    where:
      t_k = Tr(rho^k)  -- spectral moments (3 features)
      a2  = Tr(rho_A^2)  -- subsystem A purity
      b2  = Tr(rho_B^2)  -- subsystem B purity
      λ_i = sorted eigenvalues of partial transpose (4 features)
    """
    rho_A = partial_trace_A(rho)
    rho_B = partial_trace_B(rho)

    t2 = np.trace(rho @ rho).real
    t3 = np.trace(rho @ rho @ rho).real
    t4 = np.trace(rho @ rho @ rho @ rho).real
    a2 = np.trace(rho_A @ rho_A).real
    b2 = np.trace(rho_B @ rho_B).real

    rho_pt = partial_transpose_B(rho)
    pt_eigs = np.sort(np.linalg.eigvalsh(rho_pt))

    return np.array([t2, t3, t4, a2, b2, *pt_eigs])
```

```
import os

def convert_file(input_path, output_path=None):
    """
    Reads density matrices from input_path (32 numbers per line:
    Re(ρ_11), Im(ρ_11), Re(ρ_12), Im(ρ_12), ..., row-major order)
    and writes the 9 LU-invariant features per line to output_path.

    If output_path is None, automatically uses: <input_stem>_features<ext>
    e.g., 'mixedall.txt' → 'mixedall_features.txt'
    """
    if output_path is None:
        stem, ext = os.path.splitext(input_path)
        output_path = f"{stem}_features{ext}"

    n_processed = 0
    with open(input_path, 'r') as f_in, open(output_path, 'w') as f_out:
        for line in f_in:
            nums = np.array([float(x) for x in line.split()])
            if len(nums) != 32:
                continue

            complex_vals = nums[::2] + 1j * nums[1::2]
            rho = complex_vals.reshape(4, 4)

            features = lu_invariants(rho)

            f_out.write('    '.join(f'{x:.10f}' for x in features) + '\n')
            n_processed += 1

    print(f"Converted {n_processed} matrices: {input_path} → {output_path}")
    return n_processed
```

```
#!/usr/bin/env python3
"""Transpose rows of numbers into tagged columns.

Input: one row per line, numbers separated by commas (spaces are ignored).
Output: one line per column, prefixed by a tag, listing that column's value
        across every row.

Usage:
    python3 transpose_numbers.py                # paste rows, then press Ctrl-D
    python3 transpose_numbers.py data.txt       # read rows from a file

You can edit TAGS below to give the columns custom names. If there are more
columns than tags, the extra ones fall back to Tag1, Tag2, ...
"""

import sys

# Optional custom column names. Leave empty to use Tag1, Tag2, ...
TAGS = ['TO[0.0,0.1] ','TO[0.1,0.2] ','TO[0.2,0.3]','TO[0.3,0.4] ','TO[0.4,0.5] ']


def read_rows(source):
    rows = []
    for line in source:
        line = line.strip()
        if not line:
            continue
        # Split on commas, drop empty pieces, convert to float.
        values = [float(x) for x in line.split(',') if x.strip() != '']
        rows.append(values)
    return rows


def tag_for(col_index):
    if col_index < len(TAGS):
        return TAGS[col_index]
    return f'Tag{col_index + 1}'


def main():
    if len(sys.argv) > 1:
        with open(sys.argv[1]) as f:
            rows = read_rows(f)
    else:
        rows = read_rows(sys.stdin)

    if not rows:
        print('No numbers were provided.')
        return

    n_cols = max(len(r) for r in rows)
    for col in range(n_cols):
        # Pick the value of this column from every row that has it.
        column_values = [str(row[col]) for row in rows if col < len(row)]
        print(f'[{tag_for(col)}] ' + ' '.join(column_values))


if __name__ == '__main__':
    main()
```
